```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv('Teen_Mental_Health_Dataset.csv')
df['gender_code'] = df['gender'].map({'male': 0, 'female': 1})
X = df[['daily_social_media_hours', 'sleep_hours']]
y = df['stress_level']

model = DecisionTreeClassifier()
model.fit(X, y)

social_media_time_input = float(input("How much time do you spend on social media"))
sleep_hours_input = float(input("How much time do you sleep daily"))

user_data = [[social_media_time_input, sleep_hours_input]]
stress_level_prediction = model.predict(user_data)

print("Your predicted stress level is : " + str(stress_level_prediction) + "/10")

```

```python
import matplotlib.pyplot as plt

avrage_social_media_time = df.groupby('age')['daily_social_media_hours'].mean()

plt.plot(avrage_social_media_time.index, avrage_social_media_time.values, marker = "o", linewidth=3)
plt.title("Avrage social media time by age")
plt.xlabel("Age")
plt.ylabel("Hours daily")
plt.show()
plt.close()



```
